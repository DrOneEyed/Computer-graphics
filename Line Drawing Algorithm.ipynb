```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
c0 = list(map(int,input("Enter Starting Coordinates: ").split(","))) 
c1 = list(map(int,input("Enter Ending Coordinates: ").split(","))) 
```

```python
dx = abs(c1[0] - c0[0])
dy = abs(c1[1] - c0[1])
m = dy/dx
e = m - 0.5
print("dx =", dx, "\ndx =", dy, "\nSlope =", m, "\nIntersection Point =", e)
```

```python
disp_line = []
x, y = c0[0], c0[1]

for i in range(dx):
  disp_line.append([x, y])
  while(e>0):
    y=y+1
    e=e-1
  x=x+1
  e=e+m

disp_line.append(c1)
```

```python
x, y = zip(*disp_line)

plt.scatter(x, y, color='orange')
plt.plot(x, y, color = "black")
for i_x, i_y in disp_line:
    plt.text(i_x, i_y, '({}, {})'.format(i_x, i_y))

plt.show()
```
