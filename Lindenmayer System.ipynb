# Implmentation Of Lindenmayer System

### Importing LIbraries

```python
from ColabTurtle.Turtle import *
```

```python
def createLSystem(numIters,axiom):
  startString = axiom
  endString = ""
  for i in range(numIters):
    endString = processString(startString)
    startString = endString
  return endString

def processString(oldStr):
  newstr = ""
  for ch in oldStr:
    newstr = newstr + applyRules(ch)
  return newstr

def applyRules(ch):
  newstr = ""
  if ch == 'F':
    newstr = 'FF++FF--FFBB++BB--BB'
  elif ch == 'B':
    newstr = '++F--B'
  else:
    newstr = ch   
  return newstr

def drawLsystem(instructions, angle, distance):
  for cmd in instructions:
    if cmd == 'F':
      forward(distance)
    elif cmd == 'B':
      backward(distance)
    elif cmd == '+':
      right(angle)
    elif cmd == '-':
      left(angle)

def main():
  inst = createLSystem(3, "F")  
  print(inst)
  initializeTurtle()
  up()
  down()
  speed(12)
  drawLsystem(inst, 60, 12)
  hideturtle()

main()
```
