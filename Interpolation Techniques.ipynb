# ***Perform interpolation techniques such as : NN, Bi-Linear, Bi-Cubic for images***

# Importing Libs


```python
import cv2
from skimage import io
from google.colab.patches import cv2_imshow
import matplotlib.pyplot  as plt
import numpy as np
```

# ***For Gray Scale Img***

# Image Reading And Orignal Display

```python
img = io.imread('https://i.pinimg.com/736x/13/a8/ca/13a8ca44f9c620a42cbb1aa2ae149a4a.jpg') # a grayscale img
img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(img)
```

# Calculating Dimensions To Preserve The Aspect Ration Of The Image

```python
scale_percent = 220
dim = (int(img.shape[1] * scale_percent / 100), int(img.shape[0] * scale_percent / 100))
```

#Normal Interpolation

```python
resized = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
#cv2_imshow(resized)
```

# Nearest Neighbour Interpolation

```python
resized_nn = cv2.resize(img, dim, interpolation = cv2.INTER_NEAREST)
#cv2_imshow(resized_nn)
```

#Linear Interpolation

```python
resized_li = cv2.resize(img, dim, interpolation = cv2.INTER_LINEAR)
#cv2_imshow(resized_li)
```

# Cubic Interpolation

```python
resized_cu = cv2.resize(img, dim, interpolation = cv2.INTER_CUBIC)
#cv2_imshow(resized_cu)
```

# Comparison Between All The Three Interpolations

```python
cv2_imshow(cv2.hconcat([resized_nn, resized_li, resized_cu]))
print("Original Size:" + str(img.shape) + "\nAfter Interpolation:" + str(resized_nn.shape) + "\t\t\t\t\t\t" + str(resized_li.shape) + "\t\t\t\t\t\t" + str(resized_cu.shape))
```
