# Implementing Midpoint Circle Algorithm

### Importing Libraries 

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Taking Input of Starting & End Coordinates

```python
c = list(map(int,input("Enter Coordinates: ").split(",")))
r = int(input("Enter Radius: "))
```

### Midpoint Circle Drawing Algorithm

```python
quad = []

x = 0
y = r
xc, yc = c
p = 1-r

while(x<y):
  quad.append([x, y])

  if p < 0:
    x += 1
    p += 1 + 2*x
  else:
    x += 1
    y -= 1
    p += 1 - 2*y + 2*x

quad.append([x,y])

for i in range(len(quad)):
  quad.append([quad[i][1], quad[i][0]])
x_quad1, y_quad1 = zip(*quad)

x_quad2 = [(xc - x) for x in x_quad1]
y_quad2 = [(yc + y) for y in y_quad1]

x_quad3 = [(xc - x) for x in x_quad1]
y_quad3 = [(yc - y) for y in y_quad1]

x_quad4 = [(xc + x) for x in x_quad1]
y_quad4 = [(yc - y) for y in y_quad1]

x_quad1 = [(xc + x) for x in x_quad1]
y_quad1 = [(yc + y) for y in y_quad1]
```

### Plotting The Circle

```python
fig = plt.figure()
ax = fig.add_subplot(111)

plt.plot(x_quad1, y_quad1, x_quad2, y_quad2, x_quad3, y_quad3, x_quad4, y_quad4)

ax.set_aspect('equal', adjustable='box')


plt.title("Bresenham Circle Drawing Algorithm")

plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.show()
```
